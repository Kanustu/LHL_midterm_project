```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# creation of df
tornado_df = pd.read_csv('/Users/jordankanius/Downloads/tornados.csv')
```

```python
# function for seaborn plot choice looking at tornados per year
def plot_count_by_year(plot_type, x_column, y_column):
    # df grouped by year
    new_df = tornado_df.groupby('yr').count().reset_index()
    plot_type(data = new_df, x = new_df[x_column], y= new_df[y_column])
```

```python
plot_count_by_year(sns.regplot,'yr', 'om')
```

```python
plot_count_by_year(sns.lineplot, 'yr', 'om')
```
